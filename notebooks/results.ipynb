# Final PF4 Association Table

This notebook combines the processed results into a single final table to create a clean and harmonized dataset of PF4-related variants for downstream analysis and comparison across cardiovascular phenotypes.

```python
from pathlib import Path
import json
import pandas as pd

pd.set_option("display.max_columns", None)
```

```python
gwas_path = Path("../data/interim/gwas_catalog/associations.csv")
hermes_path = Path("../data/interim/hermes/associations.csv")

cardio_cad_path = Path("../data/interim/cardiogram_c4d/cad_associations.csv")
cardio_mi_path = Path("../data/interim/cardiogram_c4d/mi_associations.csv")

finngen_hf_path = Path("../data/interim/finngen/hf_associations.csv")
finngen_cad_path = Path("../data/interim/finngen/cad_associations.csv")
finngen_mi_path = Path("../data/interim/finngen/mi_associations.csv")

out_dir = Path("../data/final")
out_dir.mkdir(parents=True, exist_ok=True)

out_csv = out_dir / "pf4_final.csv"
out_summary_json = out_dir / "summary.json"
```

```python
final_columns = [
    "source_dataset",
    "phenotype",
    "trait_name",
    "association_id",
    "variant_id",
    "rsID",
    "chromosome",
    "position",
    "effect_allele",
    "other_allele",
    "effect_size",
    "standard_error",
    "odds_ratio",
    "ci_lower",
    "ci_upper",
    "p_value",
    "allele_frequency",
    "mapped_genes",
    "nearest_genes",
    "is_selected_trait",
]
```

```python
def ensure_final_columns(df):
    df = df.copy()

    for col in final_columns:
        if col not in df.columns:
            df[col] = pd.NA

    return df[final_columns]
```

```python
gwas_df = pd.read_csv(gwas_path)

gwas_final = pd.DataFrame({
    "source_dataset": "GWAS Catalog",
    "phenotype": pd.NA,
    "trait_name": gwas_df.get("trait_name"),
    "association_id": gwas_df.get("association_id"),
    "variant_id": pd.NA,
    "rsID": gwas_df.get("rsID"),
    "chromosome": pd.NA,
    "position": pd.NA,
    "effect_allele": gwas_df.get("risk_allele"),
    "other_allele": pd.NA,
    "effect_size": pd.NA,
    "standard_error": pd.NA,
    "odds_ratio": pd.NA,
    "ci_lower": pd.NA,
    "ci_upper": pd.NA,
    "p_value": gwas_df.get("p_value"),
    "allele_frequency": gwas_df.get("risk_frequency"),
    "mapped_genes": gwas_df.get("mapped_genes"),
    "nearest_genes": pd.NA,
    "is_selected_trait": gwas_df.get("is_selected_trait"),
})

gwas_final = ensure_final_columns(gwas_final)

gwas_final.head()
```

```python
hermes_df = pd.read_csv(hermes_path)

hermes_final = pd.DataFrame({
    "source_dataset": "HERMES",
    "phenotype": "Heart failure",
    "trait_name": "Heart failure",
    "association_id": pd.NA,
    "variant_id": hermes_df.get("HERMES_marker_name"),
    "rsID": hermes_df.get("rsID"),
    "chromosome": hermes_df.get("HERMES_CHR"),
    "position": hermes_df.get("HERMES_BP"),
    "effect_allele": hermes_df.get("HERMES_Allele1"),
    "other_allele": hermes_df.get("HERMES_Allele2"),
    "effect_size": hermes_df.get("HERMES_effect"),
    "standard_error": hermes_df.get("HERMES_std_error"),
    "odds_ratio": pd.NA,
    "ci_lower": pd.NA,
    "ci_upper": pd.NA,
    "p_value": hermes_df.get("HERMES_p_value"),
    "allele_frequency": hermes_df.get("HERMES_Allele1_freq"),
    "mapped_genes": pd.NA,
    "nearest_genes": pd.NA,
    "is_selected_trait": pd.NA,
})

hermes_final = ensure_final_columns(hermes_final)

hermes_final.head()
```

```python
def build_cardiogram_final(path, source_label, phenotype_label, prefix):
    df = pd.read_csv(path)

    final_df = pd.DataFrame({
        "source_dataset": source_label,
        "phenotype": phenotype_label,
        "trait_name": phenotype_label,
        "association_id": pd.NA,
        "variant_id": df.get(f"{prefix}_hm_variant_id"),
        "rsID": df.get(f"{prefix}_hm_rsid"),
        "chromosome": df.get(f"{prefix}_hm_chrom"),
        "position": df.get(f"{prefix}_hm_pos"),
        "effect_allele": df.get(f"{prefix}_hm_effect_allele"),
        "other_allele": df.get(f"{prefix}_hm_other_allele"),
        "effect_size": df.get(f"{prefix}_hm_beta"),
        "standard_error": df.get(f"{prefix}_standard_error"),
        "odds_ratio": df.get(f"{prefix}_hm_odds_ratio"),
        "ci_lower": df.get(f"{prefix}_hm_ci_lower"),
        "ci_upper": df.get(f"{prefix}_hm_ci_upper"),
        "p_value": df.get(f"{prefix}_p_value"),
        "allele_frequency": df.get(f"{prefix}_hm_effect_allele_frequency"),
        "mapped_genes": pd.NA,
        "nearest_genes": pd.NA,
        "is_selected_trait": pd.NA,
    })

    return ensure_final_columns(final_df)
```

```python
cardio_cad_final = build_cardiogram_final(
    cardio_cad_path,
    source_label="CARDIoGRAMplusC4D",
    phenotype_label="Coronary artery disease",
    prefix="CAD"
)

cardio_mi_final = build_cardiogram_final(
    cardio_mi_path,
    source_label="CARDIoGRAMplusC4D",
    phenotype_label="Myocardial infarction",
    prefix="MI"
)

cardio_cad_final.shape, cardio_mi_final.shape
```

```python
def build_finngen_final(path, phenotype_label):
    df = pd.read_csv(path)

    final_df = pd.DataFrame({
        "source_dataset": "FinnGen",
        "phenotype": phenotype_label,
        "trait_name": phenotype_label,
        "association_id": pd.NA,
        "variant_id": pd.NA,
        "rsID": df.get("FINNGEN_rsids"),
        "chromosome": df.get("FINNGEN_CHR"),
        "position": df.get("FINNGEN_BP"),
        "effect_allele": df.get("FINNGEN_alt"),
        "other_allele": df.get("FINNGEN_ref"),
        "effect_size": df.get("FINNGEN_beta"),
        "standard_error": df.get("FINNGEN_std_error"),
        "odds_ratio": pd.NA,
        "ci_lower": pd.NA,
        "ci_upper": pd.NA,
        "p_value": df.get("FINNGEN_p_value"),
        "allele_frequency": df.get("FINNGEN_alt_allele_freq"),
        "mapped_genes": pd.NA,
        "nearest_genes": df.get("FINNGEN_nearest_genes"),
        "is_selected_trait": pd.NA,
    })

    return ensure_final_columns(final_df)
```

```python
finngen_hf_final = build_finngen_final(
    finngen_hf_path,
    phenotype_label="Heart failure"
)

finngen_cad_final = build_finngen_final(
    finngen_cad_path,
    phenotype_label="Coronary artery disease"
)

finngen_mi_final = build_finngen_final(
    finngen_mi_path,
    phenotype_label="Myocardial infarction"
)

finngen_hf_final.shape, finngen_cad_final.shape, finngen_mi_final.shape
```

```python
dfs = [
    gwas_final,
    hermes_final,
    cardio_cad_final,
    cardio_mi_final,
    finngen_hf_final,
    finngen_cad_final,
    finngen_mi_final,
]

dfs = [df for df in dfs if df is not None and not df.empty]

final_df = pd.concat(dfs, ignore_index=True)

final_df = ensure_final_columns(final_df)

numeric_cols = [
    "position",
    "effect_size",
    "standard_error",
    "odds_ratio",
    "ci_lower",
    "ci_upper",
    "p_value",
    "allele_frequency",
]

for col in numeric_cols:
    final_df[col] = pd.to_numeric(final_df[col], errors="coerce")

final_df = final_df.sort_values(
    ["p_value", "source_dataset", "phenotype"],
    ascending=[True, True, True],
    na_position="last"
)

rows_before_deduplication = len(final_df)

final_df = final_df.drop_duplicates()

rows_after_deduplication = len(final_df)
duplicate_rows_removed = rows_before_deduplication - rows_after_deduplication

final_df.shape
```

```python
final_df.to_csv(out_csv, index=False)

print("Saved:", out_csv)
```

```python
summary = {
    "total_rows": int(len(final_df)),
    "unique_rsIDs": int(final_df["rsID"].dropna().nunique()),
    "duplicate_rows_removed": int(duplicate_rows_removed),
    "source_datasets": final_df["source_dataset"].dropna().astype(str).unique().tolist(),
    "phenotypes": final_df["phenotype"].dropna().astype(str).unique().tolist(),
    "rows_by_source_and_phenotype": {
        f"{source} | {phenotype}": int(count)
        for (source, phenotype), count in final_df
        .fillna({"source_dataset": "Unknown", "phenotype": "Unknown"})
        .groupby(["source_dataset", "phenotype"])
        .size()
        .to_dict()
        .items()
    },
}

with open(out_summary_json, "w") as f:
    json.dump(summary, f, indent=4)

summary
```
